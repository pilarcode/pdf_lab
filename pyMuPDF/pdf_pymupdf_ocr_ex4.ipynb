1.Download Tesseract:

Go to the Tesseract at UB Mannheim page.
Download the latest version of the Tesseract installer for Windows.

2. Install Tesseract:

Run the downloaded installer and follow the installation instructions.
During installation, make sure to select the option to add Tesseract to your system PATH.
Verify Installation:

Open a Command Prompt and type tesseract -v to verify that Tesseract is installed correctly and added to your PATH.

3. Download tessdata:

Tesseract uses language data files (tessdata) to recognize text. These files are usually included in the installation, but you can download additional language files from the tessdata repository.
Set tessdata path (if needed):

If you have custom tessdata files, you can set the TESSDATA_PREFIX environment variable to point to the directory containing your tessdata files.

```python
source = "../070ZNNLE73289_S1097.png"
```

```python
import pymupdf 


doc = pymupdf.open(source)
print(doc)
```

```python
pno =0
page = doc.load_page(pno)  # loads page number 'pno' of the document (0-based)
page = doc[pno] 
```

* get_text

```python
from pprint import pprint
d = page.get_text("dict")
d
```

* get_textpage_ocr

```python
text_page =page.get_textpage_ocr()
```

* blocks

```python
blocks =text_page.extractBLOCKS()
print(len(blocks))
blocks
```

```python
texts =text_page.extractTEXT()

from pprint import pprint
pprint(texts)
```
